```python
import pandas as pd
import numpy as np
```

```python
stagione_17 = pd.read_csv("2020-21.csv", sep=',', header=None)
```

```python
stagione_17['sky'] = np.where((stagione_17[4]=='SKY')  , 1, 0)
stagione_17['mediaset'] = np.where((stagione_17[4]=='PREMIUMMEDIASET'), 1, 0)
stagione_17['serieA_tv'] = np.where((stagione_17[4]=='SerieATIMTV'), 1, 0)
stagione_17['dazn'] = np.where((stagione_17[4]=='DAZN'), 1, 0)
stagione_17=stagione_17.drop([4], axis=1)
stagione_17=stagione_17.rename(columns = {0: 'squadra1', 1: 'squadra2', 2: 'data', 3: 'ora'}, inplace = False)
stagione_17.to_csv('stagione_21.csv', index=False) 
```
